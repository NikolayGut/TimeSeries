<h1 style="color: black;"><i>Построение моделей ARIMA, AR, MA, ARMA для стационарных временных рядов. Выбор наилучшей модели для поставленной задачи</i></h1>

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf

plt.style.use("seaborn-v0_8-deep")
plt.rcParams['figure.figsize'] = (15, 5)
```

```python
xl = pd.ExcelFile("/Users/nikolaygut/Desktop/Data.xlsx")
df = xl.parse("Sheet1")
df.head()
```

```python
data = pd.DataFrame(df[df['ts17'].isnull() == False]['ts17'])
data
```

<h3>Выведем данные в виде графика</h3>

```python
plt.plot(data.index, data, color='green');

plt.ylabel("Y")
plt.xlabel("X")
plt.title("Data")

plt.show()
```

```python
borderline_index = data.size * 90 // 100
```

<h3>Разделим выборку на треин и тест</h3>

```python
train = data[data.index <= borderline_index]
test = data[data.index >= borderline_index]
```

```python
plt.plot(train, color = "green")
plt.plot(test, color = "red")
plt.ylabel("Y")
plt.xlabel("X")
plt.title("Train/Test split for Data")
plt.show()
```

<h3>Построим график АКФ</h3>

```python
plot_acf(train, alpha = 0.05)
plt.title("АКФ для обучающей выборки")
plt.show()
```

<h3>Построим график ЧАКФ</h3>

```python
plot_pacf(train, alpha = 0.05)
plt.title("ЧАКФ для обучающей выборки")
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import warnings
```

<h3>Число значимых членов АКФ - 3, число значимых членов ЧАКФ - 3. Поэтому построим варианты ARMA моделей порядков p=[0, 3], q=[0,3]</h3>

```python
list_mod=[]
params=[(0,0,2),(2,0,0),(2,0,3),(0,0,3),(3,0,0),(3,0,3),(1,0,0),(0,0,1)]
for param in params:
    mod = SARIMAX(train, order=param, enforce_stationarity=False, enforce_invertibility=False, )
    results = mod.fit(disp=0)
    print(results.summary())
    list_mod.append((param, results.pvalues))
```

<h3>Отберем модели только со значимыми коэффициентами. То есть в столбце P>|z| значения должны быть меньше 0,05. Тогда уровень достоверности установлен на 5%, в этом случае все коэффициенты явно значимы. </h3>

```python
good_models=[]
for model in list_mod:
    flag = True
    for p in model[1]:
        if p >= 0.05:
            flag = False
    if flag:
        good_models.append(model[0])
```

```python
good_models
```

<h3>Выведем для каждой найденной модели необходимые показатели</h3>

```python
list_mod2=[]
params=good_models
for param in params:
    mod = SARIMAX(train, order=param, enforce_stationarity=False, enforce_invertibility=False)
    results = mod.fit(disp=0)
    print(results.summary())
    results.plot_diagnostics()
    list_mod2.append({'Model':'ARIMA'+str(param), 
                      'Prob(JB)': round(results.test_normality('jarquebera')[0][1], 2), 
                      'Prob(Q)': round(results.test_serial_correlation('ljungbox')[0][1][0], 2),
                      'Prob(H)': round(results.test_heteroskedasticity('breakvar')[0][1], 2),
                      'AIC': round(results.aic, 2)})
```

```python
for mod in list_mod2:
    print(mod)
```

<h3>Мы можем увидеть, что у моделей ARIMA(0, 0, 2) и ARIMA(3, 0, 0) лучше критерий Акаике. Выберем для дальнейшего анализа именно эти модели</h3>

<h3>Нулевые гипотезы для вышепредставленных Prob:<h3>
    <li>- Для prob(JB) - Остатки распределены нормально;
    <li>- Для prob(H) - Гетероскедастичность отсутствует;
    <li>- Для prob(Q) - Корреляция отсутствует.

<h3>Рассмотрим Prob значения моделей, представленные выше:</h3>

<h3>1) Для обеих моделей вероятность ошибиться, отвергнув нулевую гипотезу о нормальном распределении остатков, очень высока, более (0.4). Не отвергаем нулевую гипотезу и приходим к выводу, что остатки обеих моделей распределены нормально.</h3>

<h3>2) Вероятность ошибиться при отвержении нулевой гипотезы о корреляции остатков у обеих моделей слишком высока. Делаем вывод, что остатки не коррелируемы.</h3>

<h3>3) Не отвергаем нулевую гипотезу об отсутствии гетероскедастичности остатков модели, т.к. вероятность ошибиться большая (> 0.05). Остатки обеих моделей однородны.</h3>

<h3>Вывод: остатки у обеих моделей распределены нормально, однородны и не коррелируемы. Так как обе модели почти одинаковые по характеру остатков, сравним их, предсказав следующее значение ряда.</h3>

<h3>Предскажем следующие значения с помощью модели (0,0,2) и выведем их красным цветом, сравнив с реальными</h3>

```python
model = SARIMAX(train, order=(0, 0, 2), enforce_stationarity=False, enforce_invertibility=False)
fit_model = model.fit(disp=0)
pred1 = fit_model.get_prediction(start=len(train)-1, end=len(train)+len(test)-2, dynamic=False)
```

```python
ax = test.plot()
pred1.predicted_mean.plot(ax=ax, alpha=.7)
```

```python
plt.plot(pred1.predicted_mean, color = "red")
plt.plot(data, color = "black")
plt.ylabel("Y")
plt.xlabel("X")
plt.title("Predict")
plt.show()
```

<h3>Предскажем следующие значения с помощью модели (3,0,0) и выведем их красным цветом, сравнив с реальными</h3>

```python
model = SARIMAX(train, order=(3, 0, 0), enforce_stationarity=False, enforce_invertibility=False)
fit_model = model.fit(disp=0)
pred2 = fit_model.get_prediction(start=len(train)-1, end=len(train)+len(test)-2, dynamic=False)
```

```python
ax = test.plot()
pred2.predicted_mean.plot(ax=ax, alpha=.7)
```

```python
plt.plot(pred2.predicted_mean, color = "red")
plt.plot(data, color = "black")
plt.ylabel("Y")
plt.xlabel("X")
plt.title("Predict")
plt.show()
```
